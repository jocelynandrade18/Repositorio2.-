# **Clasificación del Iris**

```python
from sklearn.datasets import load_iris  # Importa la función para cargar el dataset Iris desde scikit-learn
iris = load_iris() # Carga el dataset y lo asigna a la variable 'iris'
```

```python
type(iris) # Para saber que valores tiene el Iris
```

```python
print(iris.keys()) # Imprime la llave del Iris
```

```python
print(iris.DESCR) #Describe lo que contiene la flor
```

```python
import pandas as pd # Convertir el conjunto de Datos a Panda
iris_df = pd.DataFrame(data=iris['data'], columns=iris['feature_names']) # Definimos las columnas en datos de un Iris
iris_df # Mostramos los DataFrame del Iris
```

```python
iris_df.describe() # Describe los elementos de las filas y columnas del Iris
```

```python
X = iris_df # En el eje X estarán el DataFrame de la Iris
y = pd.get_dummies(iris.target).values # En las Y, evalúa los datos
```

```python
from sklearn.model_selection import train_test_split # Importa el modelo

X_train, X_test, y_train, y_test = train_test_split( #
    X, y, #X conjunto de entrada y Y conjunto de salida
    test_size=0.2, random_state=42 # Utiliza solo el 20% de los Datos, y los números aleatorios de 42 elementos
)
```
