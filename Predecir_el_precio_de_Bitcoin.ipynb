```python
import pandas as pd # Importamos Pandas

_df = pd.read_csv("https://gist.githubusercontent.com/mevangelista-alvarado/6f4f28c00f9e683108637cb3c8d5db89/raw/1ce753283fce23eeab952fa1d660a7069bbe9104/Bitcoin_1_1_2024-6_9_2024_historical_data_coinmarketcap.csv", delimiter=';') # Lee el CSV de GitHub con información de Bitcoin
# Ordenamos correctamente
_df = _df.sort_index(ascending=False) # Ordena el DataFrame por índice, de mayor a menor
_df.head() # Muestra las primeras 5 filas
```

```python
'timeOpen', 'timeClose', 'timeHigh', 'timeLow', 'name', 'open', 'high',
'low', 'close', 'volume', 'marketCap', 'timestamp' #  Selecciona una lista de columnas específicas
```

```python
df = _df[['timeOpen', 'open', 'high', 'low', 'close']] # Muestra 5 de estas columnas
```

```python
dates = df[['close']].values  # Codigo para Precio
```

```python
from sklearn.preprocessing import MinMaxScaler # Importa la clase para escalar datos

scaler = MinMaxScaler(feature_range=(0, 1)) # Crea un escalador para el rango 0 a 1
scaled_data = scaler.fit_transform(dates) # Ajusta (fit) y transforma (scale) los 'dates' (precios)
```

```python
window_size = 60 # Definimos el tamaño de la ventana
```

```python
import numpy as np # Importa NumPy

def create_sequences(data, window_size): # Define una función para crear secuencias
    sequences = [] # Lista para secuencias (X)
    labels = [] # Lista para etiquetas (y)
    for i in range(len(data) - window_size): # Recorre los datos con un deslizamiento
        sequences.append(data[i:i+window_size]) # Agrega una ventana de datos
        # El precio es la primera columna
        labels.append(data[i + window_size, 0])  # Agrega el dato siguiente
    return np.array(sequences), np.array(labels) # Devuelve arrays de NumPy

# (Asumiendo que 'window_size' ya existe)
X, y = create_sequences(scaled_data, window_size) # Crea los conjuntos X (entradas) e y (salidas)
```

```python
split = int(len(X) * 0.8) # Define el punto de división (el índice del 80% de los datos)
X_train, y_train = X[:split], y[:split] # Crea el set de entrenamiento (el primer 80%)
X_test, y_test = X[split:], y[split:] # Crea el set de prueba (el 20% restante)
```

# **Modelo RNN**

```python
from tensorflow.keras.models import Sequential # Importa el contenedor del modelo (secuencial)
from tensorflow.keras.layers import LSTM, Dense, SimpleRNN # Importa los tipos de capas que usarás

model = Sequential() # Crea un modelo secuencial (una capa tras otra)
model.add(SimpleRNN(units=120, return_sequences=True, input_shape=(window_size, X_train.shape[2]))) # 1ra capa RNN (120 neuronas), pasa la secuencia a la siguiente capa
model.add(SimpleRNN(units=60, return_sequences=False)) # 2da capa RNN (60 neuronas), solo da el resultado final
model.add(Dense(units=30)) # Capa Densa (normal) de 30 neuronas
model.add(Dense(units=1)) # Capa de salida (1 neurona) para predecir un solo valor (el precio)
```

```python
from tensorflow.keras.optimizers import Adam # Importa el optimizador 'Adam'

learning_rate = 0.001 # Define la tasa de aprendizaje (qué tan rápido aprende el modelo)
adam_optimizer = Adam(learning_rate=learning_rate) # Crea una instancia del optimizador Adam con esa tasa

# Compilar el modelo
model.compile(optimizer=adam_optimizer, loss='mean_squared_error') # Configura el modelo: "optimizer" para ajustar pesos, "loss" para medir el error (MSE)

# Entrenar el modelo
model.fit(X_train, y_train, batch_size=1, epochs=10) #Entrenamos el modelo en 10 épocas con X y Y
```

```python
# Hacer predicciones
predictions = model.predict(X_test) # Obtiene las predicciones (valores de 0 a 1)

# Desescalar los datos predichos
# (Nota: El 'np.concatenate' y 'zeros' se usan si el scaler original tenía más columnas)
predictions = scaler.inverse_transform(np.concatenate((predictions, np.zeros((predictions.shape[0], 3))), axis=1))[:,0] # Convierte predicciones a precio real
y_test = scaler.inverse_transform(np.concatenate((y_test.reshape(-1, 1), np.zeros((y_test.shape[0], 3))), axis=1))[:,0] # Convierte 'y_test' a precio real para comparar
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error # Importa funciones para medir el error

mae = mean_absolute_error(y_test, predictions) # Calcula el Error Absoluto Medio (MAE)
rmse = np.sqrt(mean_squared_error(y_test, predictions)) # Calcula la Raíz del Error Cuadrático Medio (RMSE)

print(f"MAE: {mae}") # Imprime el valor del MAE
print(f"RMSE: {rmse}") # Imprime el valor del RMSE
```

```python
import matplotlib.pyplot as plt # Importa la biblioteca para graficar
import numpy as np # Importa NumPy

valid = df[split:] # Crea un DataFrame "valid" con los datos de prueba (el 20% restante)
# Reiniciar el índice del DataFrame de validación
valid = valid.reset_index(drop=True) # Reinicia el índice de "valid" (para que inicie en 0)
# Inicializar la columna de predicciones con NaN
valid['Predictions'] = np.nan   # Crea una nueva columna "Predictions" (inicialmente vacía)

# Añadir las predicciones al DataFrame de validación desde el punto donde inician las predicciones
valid.loc[window_size:, 'Predictions'] = predictions # Coloca las predicciones, desfasadas por el tamaño de la ventana
dates_valid = pd.to_datetime(valid['timeOpen']).apply(lambda x: x.strftime('%Y-%m-%d')).tolist() # Formatea las fechas (columna "timeOpen") para usarlas en el gráfico

plt.figure(figsize=(16,8)) # Crea la figura del gráfico con un tamaño de 16x8
plt.title('Modelo RNN para Predicción de Bitcoin') # Añade un título al gráfico
plt.xlabel('Fecha') # Pone etiqueta al eje X
plt.ylabel('Precio de Bitcoin (USD)') # Pone etiqueta al eje Y
plt.plot(dates_valid, valid[['close', 'Predictions']]) # Grafica los valores reales (2close") y las predicciones
plt.legend(['Valor Real', 'Predicciones'], loc='lower right') # Muestra la leyenda (en la esquina inferior derecha)
plt.xticks(rotation=90) # Rota las etiquetas del eje X 90 grados (para mejor lectura)
plt.show() # Muestra el gráfico
```

```python
# Generar secuencias para los siguientes días
future_sequences = [] # Lista para guardar las secuencias futuras
# Última secuencia de los valores que tenemos
last_sequence = X[-1]   # Toma la última secuencia de datos (la 'semilla')

days = 10 # Define cuántos días a futuro predecir
for _ in range(days): # Inicia un bucle para predecir 'days' veces
    # Predecir el siguiente valor
    next_value = model.predict(np.array([last_sequence]))[0, 0] # Predice el día siguiente (valor 0-1)

    # Actualizar la secuencia para la siguiente predicción
    # Quita el dato más antiguo ([1:]) y añade la nueva predicción ([[next_value]])
    last_sequence = np.concatenate((last_sequence[1:], [[next_value]]), axis=0)

    # Añadir la nueva secuencia a las secuencias futuras
    future_sequences.append(last_sequence) # Agrega la secuencia actualizada (ej: días 2-61) a la lista

# Convertir las secuencias futuras a un numpy y reshape para la RNN
future_sequences = np.array(future_sequences) # Convierte la lista de secuencias a un array NumPy
future_sequences = np.reshape(future_sequences, (future_sequences.shape[0], future_sequences.shape[1], 1)) # Reajusta la forma del array (para la RNN)

# Hacer predicciones para los siguientes días
future_predictions = model.predict(future_sequences) # Predice los 'days' (10) días futuros

# Desescalar los datos predichos para los siguientes días
future_predictions = scaler.inverse_transform(np.concatenate((future_predictions, np.zeros((future_predictions.shape[0], 3))), axis=1))[:,0] # Convierte las predicciones (0-1) a precio real
```

```python
# Última fecha en los datos reales
last_date = df['timeOpen'].iloc[-1] # Obtiene la última fecha (de la columna 'timeOpen')

future_dates = pd.date_range(start=last_date, periods=days)[1:] # Crea un rango de fechas futuras (ej: 10 días), omitiendo el primer día (que es 'last_date')
future_dates = future_dates.strftime('%Y-%m-%d').tolist() # Formatea las fechas (YYYY-MM-DD) y las convierte en una lista
```

**Graficar los resultados de los siguientes 30 días**

```python
plt.figure(figsize=(16,8)) # Crea una figura de gráfico (tamaño 16x8)
plt.title('Predicciones de los siguientes días') # Pone el título al gráfico
plt.xlabel('Fecha') # Pone etiqueta al eje X
plt.ylabel('Precio de Bitcoin (USD)') # Pone etiqueta al eje Y
plt.plot(dates_valid, valid[['close', 'Predictions']], label=['real', 'Predicciones']) # Grafica los datos pasados (reales y predichos)
plt.plot(future_dates, future_predictions[:-1], label='Predicciones') # Grafica las predicciones futuras
plt.legend() # Muestra la leyenda del gráfico
plt.xticks(rotation=90) # Rota las etiquetas del eje X (fechas)
plt.show() # Muestra el gráfico
```
